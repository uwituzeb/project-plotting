```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

fuel_econ = pd.read_csv('fuel_econ.csv')

print(fuel_econ.head(6))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=fuel_econ, x='comb', kde=True)
plt.title('Distribution of Combined Fuel Efficiency')
plt.xlabel('Combined Fuel Efficiency (mpg)')
plt.ylabel('Count')
plt.savefig('histogram.png')
plt.show()
```

```python
# The peak of the distribution appears to be around 22-24 mpg
# A higher number of vehicles is in the range between 16-30 mpg which means that the highest number of vehicles are likely to have this fuel efficiency
# The ranges are widely spread which suggests that different vehicles have varying fuel efficiency even though there is an average of where most are 
# found
# There are fewer vehicles above 40mpg which means that vehicles with high fuel efficiency are few
```

```python
numeric_columns = fuel_econ.select_dtypes(include=['int64', 'float64']).columns
numeric_data = fuel_econ[numeric_columns]

plt.figure(figsize=(12,10))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.savefig('heatmap.png')
plt.show()
```

```python
# The higher the fuel efficiency, the less C02 emitted and higher environmental scores
# Bigger engines tend to have lower fuel efficiency and higher emissions
```
